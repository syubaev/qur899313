# SnowballStemmer('english')

Check how working nltk.stem

1. Load part data without stop words
2. Use stammer

```python
import pandas as pd
from nltk.stem import SnowballStemmer
```

```python
train = pd.read_csv("../data/train_wo_sw.csv", nrows=100)
```

```python
print(train.shape)
train.head(3)
```

```python
stemmer = SnowballStemmer('english')

for text1 in train['question1']:
    text = text1.split()
    stemmed_words = [stemmer.stem(word) for word in text]
    text2 = " ".join(stemmed_words)
    print(text1, '\n', text2, '\n\n')
```
